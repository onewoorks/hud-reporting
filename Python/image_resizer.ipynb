```python
import os
from PIL import Image
```

```python
def resize_images_in_folder(folder_path, output_folder, max_width, max_height):
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            file_path = os.path.join(root, file)
            output_path = os.path.join(output_folder, file)

            # Check if the file is an image
            if file_path.lower().endswith(('.png', '.jpg', '.jpeg', '.gif')):
                # Open the image file
                image = Image.open(file_path)
                
                # Calculate new dimensions while preserving aspect ratio
                width, height = image.size
                aspect_ratio = width / height

                new_width = min(width, max_width)
                new_height = int(new_width / aspect_ratio)
                if new_height > max_height:
                    new_height = max_height
                    new_width = int(new_height * aspect_ratio)
                
                # Resize the image
                resized_image = image.resize((new_width, new_height))
                
                # Save the resized image to the output folder
                resized_image.save(output_path)
                print(f"Resized: {file_path} -> {output_path}")
                
def read_images_in_folder(folder_path):
    output_folder = '/Users/iwang/Documents/Development/Projects/hud/reporting/storage/output/KEMAMAN_01'
    image_files = []
    file_no = 0
    for root, dirs, files in os.walk(folder_path):
        for i,file in enumerate(files):
            file_path = os.path.join(root, file)
            
            filename, extension = os.path.splitext(file)
            new_filename = f"{i+1}{extension}"
            output_path = os.path.join(output_folder, new_filename)
            print(output_path)
#             # Check if the file is an image
#             if file_path.lower().endswith(('.png', '.jpg', '.jpeg', '.gif')):
#                 # Open the image file
#                 image = Image.open(file_path)
#                 image_files.append((file_path, image))
    
    return image_files
```
